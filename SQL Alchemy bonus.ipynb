```python
import pandas as pd 
import matplotlib.pyplot as plt
from sqlalchemy import create_engine 
engine = create_engine('postgresql://{user:pass}@localhost:5432/EmployeeSQL') #username and password redacted
connection = engine.connect()
```

```python
emp_data = pd.read_sql("SELECT * FROM merge4", connection)
emp_data
```

```python
#Histogram of Salary Ranges 
plt.hist(emp_data['salary'])
plt.xlabel('Salaries')
plt.ylabel('Employee Numbers')
```

```python
#bar chart of average salary by title 

salary_range = emp_data.groupby(['title'])['salary'].mean()
salary_chart = salary_range.plot(kind='bar')
salary_chart.set_xlabel('Employee Title')
salary_chart.set_ylabel('Average Salary')
plt.show()
```
